#### What's the mean `Salary` by `Department`?
**Tip**: There are 2 ways to do it: with or without pivot tables. Do it both ways. 

```python
# Method 1
employees[['Salary','Department']].groupby('Department').mean()
```

```python
# Method 2
employees.pivot_table(columns='Department', values='Salary')
```

#### Now, group the data by more than one column. Use Pandas pivot table function to select the mean `Salary` by `Department` and `Title`.

```python
employees.pivot_table(columns='Title', index = 'Department', values='Salary')
```

#### Display the number of `Years` of work experience and the `Salary` for each `Title`. Add `Years` as a column in your pivot table.
**Hint**: Use Pandas documentation to explore the [pivot table](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) function.

```python
employees.pivot_table(columns='Years', index = 'Title', values='Salary')
```

#### Find the mean `Salary` and number of employees in each `Department` and for each job `Title`.

```python
employees.pivot_table(index = ['Department','Title'], values=['Salary'], aggfunc=['mean', 'count'])
```

#### For each `Department` and each job `Title`, find the `Salary` median, minimum, maximum and standard deviation. Comment your results.

```python
employees.pivot_table(index = ['Department','Title'], values=['Salary'], aggfunc=['min',max, 'median','std'])
```

```python
"""
For an associate in the IT department, the minimum salary is actually the highest amongst other associates in other
departments. However, the maximum salary is lowest when we do the same comparison. This leads also to the lowest median
amongst all titles in all departments.
"""
```

#### The stake holders want to know, for each `Department`, the number of employees and how much money is spent on salaries. Could you provide that information? 

```python
employees.pivot_table(index = ['Department'], values=['Salary'], aggfunc=['sum', 'count'])
```

#### For each `Department` and `Title`, find the the total years of work experience and the mean `Salary`.

```python
employees.pivot_table(index = ['Department','Title'], values=['Salary','Years'], aggfunc={'Salary': 'mean', 'Years': 'sum'})
```

## Bonus Challenge 1

#### Compute the mean `Salary` after removing the maximum `Salary` for each `Department`. Create the appropriate pivot table.
**Hint**: Write a custom function.

```python
def mean_without_max(x):
    return x.sort_values()[:-1].mean() #Ordena o salário de forma ascendente, e o -1 não considera o último valor.
employees.pivot_table(index='Department',values='Salary',aggfunc=mean_without_max)
```

```python
employees[employees['Years'] > 5]
```
