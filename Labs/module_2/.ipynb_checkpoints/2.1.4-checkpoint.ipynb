````python
def mean_without_max(x):
    maax=x.max()
    
    
    return x[x!=maax].mean()
````

## Where

````python
df[df.Years>5]
````

````python
df.loc[df.Years>5]
````

````python
df[df.Years>5]['Salary']

#select df -> the whole dataframe
#select df -> the whole dataframe
  #select column Years from df -> just 1 column 
  #compare the vector with 5 ->boolean series, 1 column
#select the series of booleans from df -> all the columns, some rows
#select salary from previous step -> 1 column and some rows

#called df 2 times
````

````python
df.loc[df.Years>5]['Salary']

#it is the same, but slightly faster
#called df twice
````

````python
df.loc[df.Years>5, 'Salary']
#call df -> the whole dataframe
#call df -> the whole dataframe
  #select column Years -> just 1 column 
  #compare the vector with 5 -> boolean series, 1 column
  #I also just select column Salary -> boolean is applied to the column Salary, so I obtain 1 column and few rows
#select the specified rows and clumns from df.

#called df twice
# but much faster, since I select the column directly
````
