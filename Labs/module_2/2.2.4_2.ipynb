```python
X=df.drop(['State'], axis=1).copy()
```

```python
dct={'No':0, 'Yes':1}
```

```python
X['International plan']=X['International plan'].map(dct)
```

```python
X['Voice mail plan']=X['Voice mail plan'].map(dct)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
tsne=TSNE(random_state=11)
X_tsne=tsne.fit_transform(X_scaled)
```

```python
plt.scatter(X_tsne[:, 0],X_tsne[:, 1], c=df.Churn.map({False:'green', True:'red'}))
```

```python
plt.scatter(X_tsne[:, 0],X_tsne[:, 1], c=df['International plan'].map({'No':'green', 'Yes':'red'}))
```

```python
plt.scatter(X_tsne[:, 0],X_tsne[:, 1], c=df['Voice mail plan'].map({'No':'green', 'Yes':'red'}))
```
