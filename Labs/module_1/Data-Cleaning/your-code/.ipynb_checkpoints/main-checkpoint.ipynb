# Data Cleaning 

#### 1. Import pandas library.

```python
import pandas as pd
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data. 


```python
import pymysql
from sqlalchemy import create_engine
from getpass import getpass
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats).

```python
Hostname='relational.fit.cvut.cz'
un='guest'
password=getpass()
```

```python
db = 'stats'
```

```python
engine=create_engine(f'mysql+pymysql://{un}:{password}@{Hostname}/{db}')
```

#### 4. Import the users table.

```python
users = pd.read_sql_query("""SELECT * FROM users""", engine)
users
```

#### 5. Rename Id column to userId.

```python
users.rename(columns={'Id':'userId'}, inplace=True)
users
```

#### 6. Import the posts table. 

```python
posts = pd.read_sql_query("""SELECT * FROM posts""", engine)
posts.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
posts.rename(columns={'Id':'postId','OwnerUserId':'userId'}, inplace=True)
posts
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
new_users = users[['userId','Reputation','Views','UpVotes','DownVotes']]
new_posts = posts[['postId','Score','userId','ViewCount','CommentCount']]
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
df_merge = new_users.merge(new_posts, how='inner', on='userId')
df_merge
```

#### 10. How many missing values do you have in your merged dataframe? On which columns?

```python
df_merge.isna().sum()
```

#### 11. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before going to the next step.

```python
# As we can see in the results above, more than 50% of the rows in the column ViewCount have missing values.
df_merge.ViewCount.min()
```

```python
# The minimum value of this column is 1, which means that the posts that didn't get any views are getting a missing value instead of zero.
# So we can assume that the missing values are actually 0 views. So we will replace all the missing values with 0
df_merge.loc[df_merge.ViewCount.isna(), 'ViewCount'] = 0
df_merge
```

#### 12. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
# the ViewCount should be changed to an integer.
import numpy as np
df_merge = df_merge.astype({'ViewCount':'int64'})
df_merge.info()
```
