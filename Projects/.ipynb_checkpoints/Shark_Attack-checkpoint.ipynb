```python
# Drop columns Unnamed: 22 and Unnamed: 23 because they only have missing values, except for one row, that doesn't give more information
sharks = sharks.drop(columns='Unnamed: 22')
sharks = sharks.drop(columns='Unnamed: 23')
```

```python
# Remove rows where all values are missing
sharks = sharks.dropna(axis=0,how='all')
```

```python
# There are rows that have missing values but were not dropped because the Case Number is 0. So those rows will also be dropped
sharks.loc[sharks['Case Number'] == '0']
sharks = sharks.drop(sharks[(sharks['Case Number'] == '0') | (sharks['Case Number'] == 'xx')].index)
```

```python
# There are no duplicates in this dataset
sharks.loc[sharks.duplicated()]
```

```python
sharks = sharks.drop(columns='href formula')
sharks = sharks.drop(columns='href')
sharks = sharks.drop(columns='pdf')
```

```python
# Apparently the types "Boating" and "Boat" mean the same activity. But they started using the term "Boating" more recently. So we're going to change "Boating" to "Boat"
sharks.loc[(sharks['Type'] == 'Boating') | (sharks['Type'] == 'Boat')]
```

```python
sharks.Type = sharks.Type.replace(['Boating','Boatomg'],'Boat')
```
