# 1. Import Audio

```python
import requests as r
```

```python
# Stream = True: Significa que a ligação com o servidor mantém-se ativa até o ficheiro ser descarregado
response = r.get('https://freesound.org/data/previews/516/516531_11204368-lq.mp3', stream = True)
```

```python
# Para gravar o ficheiro no computador. 'wb' significa a forma como ler o ficheiro. w significa Write e b significa Byte.
with open('new_audio_bcn.mp3', 'wb') as f:
    f.write(response.content)
```
